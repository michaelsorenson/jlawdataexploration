```python
movies = []
for i, film in enumerate(jlaw.filmography):
    print ('{:2.1%}'.format(i / len(jlaw.filmography)), end='\r')
    try:
        movie = imdb.get_title(film)
        if movie.genres and len(movie.genres) > 0:
            movies.append(movie)
    except:
        print('failed ' + str(i))
```

```python
moviegenres = []
for movie in movies:
    moviegenres.append(movie.genres)
```

```python
filteris = []
for i, ent in enumerate(moviegenres):
    if ent:
        filteris.append(i)
```

```python
fedmovies = pd.Series(movies)[filteris].tolist()
fedgenres = [movie.genres for movie in fedmovies]
```

```python
genreCounts = defaultdict(int)
for genres in fedgenres:
    for genre in genres:
        genreCounts[genre] += 1
counts = pd.Series(genreCounts).sort_values(ascending=False)[:10]
```

```python
categories = list(counts.index)
catdict = {category: {} for category in categories}
for movie in fedmovies:
    for category in categories:
        if category in movie.genres:
            if movie.title not in catdict[category]:
                catdict[category][movie.title] = {}
            catdict[category][movie.title]['value'] = movie.rating_count
            catdict[category][movie.title]['avg'] = movie.rating
with open('../data/category_data.js', 'w') as outfile:
    outstring = "const catdata = " + json.dumps(catdict)
    outfile.write(outstring)
```

```python
# Let's experiment with Terminator
imdb_id = hg_id

# Fetch the box office numbers
base = 'https://www.boxofficemojo.com'
url = base + '/title/' + imdb_id
source = urllib.request.urlopen(url).read()
soup = BeautifulSoup(source,'lxml')
bycountry = soup

table = soup('th', text=re.compile(r'Release Group'))[0].parent.parent
group = table.findAll('tr', recursive=False)[1].find('a').get('href')
url = base + group

# Get total earnings domestic and international
source = urllib.request.urlopen(url).read()
soup = BeautifulSoup(source,'lxml')
earnings = soup('h2', text=re.compile(r'Rollout'))[0].parent.parent.findAll('div')
worldwide = earnings[1]
domestic = earnings[1].find('span', {'class': 'money'}).get_text()
domestic_url = earnings[1].find('a').get('href')
international = earnings[2].find('span', {'class': 'money'}).get_text()
worldwide = earnings[3].find('span', {'class': 'money'}).get_text()

# Get weekly domestic earnings
url = base + domestic_url
url = url[:url.rfind('/')] + '/weekly/'
source = urllib.request.urlopen(url).read()
soup = BeautifulSoup(source,'lxml')
table = soup.find('div', {'class':'a-section imdb-scroll-table-inner'}).findAll('tr')
weekly = []
for tr in table[1:]:
    date = tr.findAll('td')[0].get_text()
    earning = tr.findAll('td')[2].get_text()
    weekly.append((date, earning))
budgetdict = {
    'Total Earnings': int(worldwide.translate(str.maketrans('', '', string.punctuation))),
    'Total International': int(international.translate(str.maketrans('', '', string.punctuation))),
    'Total Domestic': int(domestic.translate(str.maketrans('', '', string.punctuation)))
}
# Print the values we've just got!
print("Total Domestic Earnings: %s" % domestic)
print("Total International Earnings: %s" % international)
print("Weekly Domestic Earnings:")
for date, earning in weekly:
    print("\t%s \t: %s" % (date, earning))
    budgetdict['Week of ' + date] = int(earning.translate(str.maketrans('', '', string.punctuation)))
with open('../data/budget_data.js', 'w') as outfile:
    outstring = "const budgetData = " + json.dumps(budgetdict)
    outfile.write(outstring)
```
